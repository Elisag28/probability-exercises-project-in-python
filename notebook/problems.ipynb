## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
# TODO
import numpy as np
import random

def probability_dices(x):
    #run the experiment x times
    times = x

    #score total
    total = 0

    for i in range(times):
        # Throw two dices x times
        dice_1 = random.randint(1,6)
        dice_2 = random.randint(1,6)

        #see if it 7 or even
        sum_dices = dice_1 + dice_2

        if (sum_dices > 7) or (sum_dices % 2 == 0):
            total += 1 
    # the total divided by the times
    return total/times

print(f"the probability that the total score is even or greater than 7 is: {np.round(probability_dices(1000) * 100,2 )} %")
```

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. All are the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"


# TODO

def take_the_ball(x):
    # x is the times that will run the experiment 
    times = x
    #the probability we need to know
    probab_1 = 0
    probab_2 = 0
    
    for i in range(x):
        ball_colors = []

        for i in range(5):
       
            ball_colors.append(ball_box[random.randint(1, 59)])

        ball_colors = np.array(ball_colors)

        whites = sum(ball_colors == 'White')
        reds = sum(ball_colors == 'Red')
        greens = sum(ball_colors == 'Green')

        if (whites == 3) and (reds == 2):
            probab_1 += 1
        if (whites == 5) or (reds == 5) or (greens == 5):
            probab_2 += 1

    return probab_1/times, probab_2/times

result = take_the_ball(1000)
print(f"the probability that we take 3 white balls and 2 reds are: {np.round(result[0] * 100, 3)} %")
print(f"the probability that we take 5 balls the same color are: {np.round(result[1] * 100, 3)} %")
```
